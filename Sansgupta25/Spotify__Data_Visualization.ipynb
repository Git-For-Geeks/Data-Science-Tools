**Correlation Heatmap Between Variable**

```python
corr_df=df.drop(["key", "mode", "explicit"],axis =1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap (corr_df,annot=True, vmin=1, vmax=1, cmap="Blues", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels (heatmap.get_xticklabels(), rotation=90)
```

Once the code had run, we had our correlation map. You may see a scale from -1 to +1 on the right side. The variables with the least or negative correlation are shown by the number -1 in this case, whereas the variables with a positive correlation are indicated by values above 0.0.

**Loudness vs Energy Correlation**

```python
plt.figure(figsize=(18,6))

sns.regplot(data=df, y="loudness", x = "energy" ,color = 'c').set(title = "Loudness vs Energy Correlation")
```

The outcome is plotted. Loudness and energy are strongly correlated with one another. You can also observe that the music and the statistics all point in the same direction. The music gets louder when the energy is higher; conversely, if the song is quieter, the energy of the track also gets quieter.

**Popularity vs Acousticness Correlation**

```python
plt.figure(figsize=(10,6))

sns.regplot(data=df, y= "popularity", x="acousticness", color ="b").set(title = "Popularity vs Acousticness Correlation")
```

Here, we can see that the blue colour regression line is pointing downward, which indicates that as a song's acoustic quality grows, its popularity declines, and vice versa as a song's popularity rises.

**With the use of a horizontal barplot, plot the song duration in relation to the various genres.**

**Duration of the Songs in Different Genres**

```python
plt.figure(figsize=(14,13))
plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap= True) 
sns.barplot(y="genre", x='duration_ms', data=df)
plt.xlabel("Duration in milli seconds") 
plt.ylabel("Genres")
```

Here, the Y-axis lists the genres, while the X-axis lists the duration in milliseconds. By analysing the data, we may determine that songs in the classical and international genres are lengthier in length than those in the children's music category.

**Top 5 Genres by Popularity**

```python
sns.set_style(style = "whitegrid") 
plt.figure(figsize=(10,5))
df=df.sort_values("popularity", ascending = False).head(10)

sns.barplot(y="genre", x='popularity', data = df).set(title= "Top 5 Genres by Popularity")
```

Here are the top 5 musical genres according to popularity: dance, pop, rap, hip-hop, and reggaeton.

**Conclusion**

It looks like the most popular songs are slightly more danceable and feature more valence. They also have zero instrumentality and little liveness.
